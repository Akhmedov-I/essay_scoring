```python
import nltk
```

```python
import docx2txt

result = docx2txt.process('/Users/islombekakhmedov/Documents/Programming training/data_science/text_identifier/essay.docx')
```

```python
tokens = nltk.word_tokenize(result)
text = nltk.Text(tokens)
tags = nltk.pos_tag(text)
```

```python
from collections import Counter
counts = Counter(tag for word,tag in tags)
print(counts)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
labels = []
sizes = []

for x, y in counts.items():
    labels.append(x)
    sizes.append(y)

fig, ax = plt.subplots(figsize=(6, 3), subplot_kw=dict(aspect="equal"))

def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:.1f}%\n({:d})".format(pct, absolute)

wedges, texts, autotexts = ax.pie(sizes, labels = labels, autopct=lambda pct: func(pct, sizes),
                                  textprops=dict(color="w"), radius = 5)

# Plot
ax.legend(wedges, labels,
          title="Ingredients",
          loc="center left",
          bbox_to_anchor=(4, 1, 5, 1))

plt.setp(autotexts, size=8, weight="bold")

ax.set_title("Words type in sentences")

plt.show()
```
